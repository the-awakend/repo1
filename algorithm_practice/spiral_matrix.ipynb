##### 2、普通算法

```python
def calculateLayer(rows):
    # 递归计算层数
    if rows ==1:
        return 1
    else:
        if rows % 2 == 0:
            # 偶数行
            return calculateLayer(rows - 1)
        else:
            return calculateLayer(rows - 1) + 1

def calCoordinate(nums):
    # 计算每一层的坐标,存入一个列表,坐标从0开始
    result = []
    for i in range(nums):
        temp = (0+i, 0+i)
        result.append(temp)
    return result

def printLayers(array, n, coordinate):
    # 从起始坐标开始打印每一层的数据
    result = []
    for num in range(len(coordinate)):
#         print(array[num][num:n-num])
#         print(array[num:n-num][n-num])
#         print(array[n-num][num:n-num][::-1])
#         print(array[num:n-num][num][::-1])       
        one = array[num][num:n-num]
        two = [i[n-num-1] for i in array[num+1:n-num]]
        three = array[n-num-1][num:n-num-1][::-1]
        four = [i[num] for i in array[num+1:n-num-1]][::-1]
        temp = [one, two, three, four]
        result = result + temp
    length = len(result)
    while result[length-1] == []:
        del(result[length-1])
        length -= 1
    return result

def main():
    n = int(input('n:'))
    rows = n
    array = []
    k = 0
    for i in range(n):
        t = []
        for j in range(n):
            k = k+1
            t.append(k)
        array.append(t)
#     print(array)
    layers = calculateLayer(rows)
#     print(layers)
    coordinates = calCoordinate(layers)
#     print(coordinates)
    result = printLayers(array, n, coordinates)
    print(result)
main()
```
