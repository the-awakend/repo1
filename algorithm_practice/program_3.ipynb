```python
## # idea应该是P行4列，每行的特征有PM序号，提出时间，优先级，所需时间,ideas的下标代表序号
# idea使用列表存储，里面有五个元素，每个元素都是一个字典
class PM():
    def __init__(self, order):
        self.order = order
        self.ideas = []
    def append_idea(self, idea):
        self.ideas.append(idea)
    def del_idea(self, idea):
        self.ideas.remove(idea)
    def most_wanted_idea(self):
        if len(self.ideas) ==0:
            return None
        elif len(self.ideas) == 1:
            return self.ideas[0]
        else:
            result_idea = self.ideas[0]
            for idea in self.ideas[1:]:
                if idea.prior < result_idea.prior:
                    result_idea = idea
                elif idea.prior == result_idea.prior:
                    if idea.time_need < result_idea.time_need:
                        result_idea = idea
                    elif idea.time_need == result_idea.time_need:
                        if idea.time_pro < result_idea.time_pro:
                            result_idea = idea
        return result_idea    
        
class Idea():
    def __init__(self, PM_order, time_pro, prior, time_need):
        self.PM_order = PM_order
        self.time_pro = time_pro
        self.prior = prior
        self.time_need = time_need
        self.finish_time = 0

def best_idea(candidates):
    if len(candidates) == 0:
        return None
    if len(candidates) == 1:
        cur_idea = candidates[0]
    else:
        cur_idea = candidates[0]
        for idea in candidates[1:]:
            if idea.time_need < cur_idea.time_need :
                cur_idea = idea
            elif idea.time_need == cur_idea.time_need:
                if idea.PM_order < cur_idea.PM_order:
                    cur_idea = idea
    return cur_idea

def do_task(idea, programmers):
    # 选取空闲的程序员或者最早完成上个任务的程序员
    if len(programmers) == 1:
        p = 0
    else:
        p = 0
        for i in range(len(programmers))[1:]:
            if programmers[i] < programmers[p]:
                p = i
    # p = i 表示选取第i个程序员
    if idea.time_pro <= programmers[p]:
        # 只有当当前idea的提出时间在程序员工作时间之前的idea才能被运算
        programmers[p] += idea.time_need # 使用第p个程序员运算idea,得到运行结束的时间。
    else:
        programmers[p] = idea.time_pro + idea.time_need
    return programmers[p]


def main():
    N, M, P = map(int, input('请分别输入产品经理数量、程序员数量、idea数量，逗号分割').split(','))
    ideas = [] # 存储所有的idea
    PM_array = [PM(i) for i in range(N+1)[1:]] # 存储所有的产品经理，PM_array[0]存储一号经理
    print('产品经理', PM_array)
    programmers = [0] * M   
    # 下标代表程序员的order，programmers[i] 的值表示，程序员i的工作时间，
    for i in range(P):
        PM_order, time_pro, prior, time_need = map(int, input('请输入idea%d的PM序号，提出时间，优先级，所需时间.以逗号分割'%i).split(','))
        idea = Idea(PM_order, time_pro, prior, time_need)
        ideas.append(idea)
        PM_array[PM_order-1].append_idea(idea)   
    print('所有的idea:')
    for k in ideas:
        print(k.PM_order, k.time_pro, k.prior, k.time_need)
    
    while True: # 开始
        # 寻找候选idea
        candidates = []
        for i in range(N):
            candidates.append(PM_array[i].most_wanted_idea())
#         print("candidates",candidates)
        for i in range(len(candidates))[::-1]:
            # 过滤掉candidates中的None
            if candidates[i] == None:
                del candidates[i]
#         print(candidates)       
         # 从cndidates中挑选一个time_need最小的idea

        best = best_idea(candidates)
        if best == None:
            break
        print('所选择的idea:', best.PM_order, best.time_pro, best.prior, best.time_need)  
        finish_time = do_task(best, programmers)
        best.finish_time = finish_time
        # 从原始的列表中删除这个idea
#         ideas.remove(idea)
        order = best.PM_order-1
        PM_array[order].del_idea(best)
    print('每个idea的完成时间:')        
    for idea in ideas:
        print(idea.finish_time)
        
main()
```
