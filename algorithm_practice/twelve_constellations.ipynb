```python
import numpy as np
from matplotlib import pyplot as plt

def simulate():
    # 一次模拟，即做一次实验，即撒满十二个星座所需要的人数，每次得到的结果都不同(一次实验要撒满十二个星座)
    count = 0
    boyfriends = np.zeros([12,1])
    while(boyfriends.sum() < 12):
        # 随机撒到某个星座上，直到撒满十二个星座
        boyfriends[np.random.randint(0,12)] = 1 
        count += 1 # 一次模拟实验得到的计数
    return count

record = np.zeros([100000,1]) # record的下标代表男友数量，值代表得到该数量进行的模拟次数

for i in range(100000):
    # 十万次模拟
    count = simulate() # 模拟实验得到count个人数
    if count >= 100000:
        count = 100000-1 # 这是为了防止越界
    record[count] += 1 # 得到count个人数的试验次数（这里有个问题，假如count>100000,从而越界了呢？）

for i in range(100000):
    record[i] /= 100000 # 得到i个人数的实验次数初一总实验次数，即得到得到该人数的概率

result = record.argmax() # 模拟次数最多的下标，即男友数量

fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.linspace(1, 100, 100), record[0:100,0], linewidth = 2)
plt.xlabel('Number of ex-boyfriends')
plt.ylabel('Probability')
plt.show()
```
